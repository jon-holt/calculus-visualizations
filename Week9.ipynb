# Figures in the xy-plane transformed on a surface plot

```
N=100
x=np.zeros([N, N])
y=np.zeros([N, N])
z=np.zeros([N, N])
zhor=np.zeros([N, N])
xcirc1=np.zeros(N)
ycirc1=np.zeros(N)
zcirc1=np.zeros(N)
xcirc2=np.zeros(N)
ycirc2=np.zeros(N)
zcirc2=np.zeros(N)
zch=np.zeros(N)
c=np.ones([N,N])*30
c[0,0]=2
c[N-1,N-1]=50

n=25
xsmile1=np.zeros(n)
ysmile1=np.zeros(n)
zsmile1=np.zeros(n)
xeye11=np.zeros(n)
yeye11=np.zeros(n)
zeye11=np.zeros(n)
xeye12=np.zeros(n)
yeye12=np.zeros(n)
zeye12=np.zeros(n)
xsmile2=np.zeros(n)
ysmile2=np.zeros(n)
zsmile2=np.zeros(n)
xeye21=np.zeros(n)
yeye21=np.zeros(n)
zeye21=np.zeros(n)
xeye22=np.zeros(n)
yeye22=np.zeros(n)
zeye22=np.zeros(n)

for i in range(N):
  for j in range(N):
    x[i,j]=i/N*4*np.pi-2*np.pi
    y[i,j]=j/N*11-5.5
    z[i,j]=4+np.sin(x[i,j])*np.arctan(y[i,j])

fig = go.Figure()

# xy-plane
fig.add_trace(go.Surface(x=x, y=y, z=z, showscale=False, visible="legendonly"))

rad=2
shift1x=-1.3
shift2x=2.5
shift2y=2.5

for i in range(N):
  xcirc1[i]=shift1x+rad*np.cos(i/N*2*np.pi)
  ycirc1[i]=rad*np.sin((i/N)*2*np.pi)
  zcirc1[i]=4+np.sin(xcirc1[i])*np.arctan(ycirc1[i])
  xcirc2[i]=xcirc1[i]-shift1x+shift2x
  ycirc2[i]=ycirc1[i]+shift2y
  zcirc2[i]=4+np.sin(xcirc2[i])*np.arctan(ycirc2[i])

for i in range(n):
        xsmile1[i]=shift1x+.5*rad*np.cos((i/N)*np.pi*2)
        ysmile1[i]=.5*rad*np.sin((i/N)*np.pi*2)
        zsmile1[i]=4+np.sin(xsmile1[i])*np.arctan(ysmile1[i])
        xeye11[i]=shift1x-.5*rad+.125*np.cos(4*i/(N)*np.pi*2)
        yeye11[i]=.125*np.sin((4*i/N)*np.pi*2)
        zeye11[i]=4+np.sin(xeye11[i])*np.arctan(yeye11[i])
        xeye12[i]=shift1x+.125*np.cos((4*i/N)*np.pi*2)
        yeye12[i]=-.5*rad+.125*np.sin((4*i/N)*np.pi*2)
        zeye12[i]=4+np.sin(xeye12[i])*np.arctan(yeye12[i])
        xsmile2[i]=shift2x+.5*rad*np.cos(np.pi/2+(i/N)*np.pi*2)
        ysmile2[i]=shift2y+.5*rad*np.sin(np.pi/2+(i/N)*np.pi*2)
        zsmile2[i]=4+np.sin(xsmile2[i])*np.arctan(ysmile2[i])
        xeye21[i]=shift2x+.125*np.cos(4*i/(N)*np.pi*2)
        yeye21[i]=shift2y-.5*rad+.125*np.sin((4*i/N)*np.pi*2)
        zeye21[i]=4+np.sin(xeye21[i])*np.arctan(yeye21[i])
        xeye22[i]=shift2x+.5*rad+.125*np.cos((4*i/N)*np.pi*2)
        yeye22[i]=shift2y+.125*np.sin((4*i/N)*np.pi*2)
        zeye22[i]=4+np.sin(xeye22[i])*np.arctan(yeye22[i])

# first face on plane
fig.add_trace(go.Scatter3d(x=xcirc1, y=ycirc1, z=zch, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xsmile1, y=ysmile1, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xeye11, y=yeye11, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xeye12, y=yeye12, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))

# first face on surface
fig.add_trace(go.Scatter3d(x=xcirc1, y=ycirc1, z=zcirc1, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xsmile1, y=ysmile1, z=zsmile1, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xeye11, y=yeye11, z=zeye11, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))
fig.add_trace(go.Scatter3d(x=xeye12, y=yeye12, z=zeye12, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='blue', width=4)))

# second face on plane
fig.add_trace(go.Scatter3d(x=xcirc2, y=ycirc2, z=zch, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xsmile2, y=ysmile2, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xeye21, y=yeye21, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xeye22, y=yeye22, z=zch[0:n], mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))

# second face on surface
fig.add_trace(go.Scatter3d(x=xcirc2, y=ycirc2, z=zcirc2, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xsmile2, y=ysmile2, z=zsmile2, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xeye21, y=yeye21, z=zeye21, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))
fig.add_trace(go.Scatter3d(x=xeye22, y=yeye22, z=zeye22, mode='lines', showlegend=False, visible="legendonly",
                                    line=dict(color='green', width=4)))

fig.update_layout(template="seaborn", 
                  scene = dict(aspectratio=dict(x=1, y=1, z=0.5),
                               camera=dict(projection=dict(type='orthographic')),
                  xaxis=dict(range=[-6,6]),
                  yaxis=dict(range=[-6,6]),
                  zaxis=dict(range=[0, 6])),
    updatemenus=[
        dict(
            type="buttons",
            active=0,
            buttons=list([
                dict(label="Surface",
                     method="update",
                     args=[{"visible": [True,
                                        False, False, False, False, 
                                        False, False, False, False, False, False, False, False, False, False, False, False]}]),
                dict(label="Face 1 on plane",
                     method="update",
                     args=[{"visible": [True,
                                        True, True, True, True, 
                                        False, False, False, False, False, False, False, False, False, False, False, False]}]),
                dict(label="Face 1 on surface",
                     method="update",
                     args=[{"visible": [True,
                                        True, True, True, True, 
                                        True, True, True, True, False, False, False, False, False, False, False, False]}]),
                dict(label="Face 2 on plane",
                     method="update",
                     args=[{"visible": [True,
                                        True, True, True, True, 
                                        True, True, True, True, True, True, True, True, False, False, False, False]}]),
                dict(label="Face 2 on surface",
                     method="update",
                     args=[{"visible": [True,
                                        True, True, True, True, 
                                        True, True, True, True, True, True, True, True, True, True, True, True]}]),
                dict(label="Faces only",
                     method="update",
                     args=[{"visible": [False,
                                        True, True, True, True, 
                                        True, True, True, True, True, True, True, True, True, True, True, True]}]),
            ]),
        )
    ])

fig.show()

fig.write_html("week9_figures_on_surface.html")
```
