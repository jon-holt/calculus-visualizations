```
import plotly.graph_objects as go
import plotly as plotly
import pandas as pd
import numpy as np
```

# Two different parameterizations of the same curve

```
N=40 # number of intervals for each variable

x1=np.zeros(N)
x2=np.zeros(N)
y1=np.zeros(N)
y2=np.zeros(N)

# data for curve
for i in range(N):
  x1[i] = 3-2*(1-2*i/(3*N))*np.cos(5*np.pi*i/(4*N))
  y1[i] = 1+1.5*(1-2*i/(3*N))*np.sin(5*np.pi*i/(4*N))
  y2[i] = 1+1.5*(1-(2/3)*(1-np.cos(np.pi*i/(2*N))))*np.sin((5*np.pi/4)*(1-np.cos(np.pi*i/(2*N))))

xmin = min(x1)
xmax = max(x1)
ymin = min(y1)
ymax = max(y1)

for i in range(N):
  x2[i] = xmax+2+3-2*(1-(2/3)*(1-np.cos(np.pi*i/(2*N))))*np.cos((5*np.pi/4)*(1-np.cos(np.pi*i/(2*N))))

# data for little line segments
lilbk1x=np.zeros(2*N)
lilbk1y=np.zeros(2*N)
lilbk2x=np.zeros(2*N)
lilbk2y=np.zeros(2*N)

for k in range(N-1): #isl+1 == k 
  lilbk1x[2*k]=x1[k];
  lilbk1x[2*k+1]=x1[k+1];
  lilbk2x[2*k]=x2[k];
  lilbk2x[2*k+1]=x2[k+1];
  lilbk1y[2*k]=y1[k];
  lilbk1y[2*k+1]=y1[k+1];
  lilbk2y[2*k]=y2[k];
  lilbk2y[2*k+1]=y2[k+1];

# plot the first curve
fig=go.Figure()
fig.add_trace(go.Scatter(x=x1, y=y1, mode='lines', showlegend=False,
                                    line=dict(
                                        color='blue',
                                        width=8)))
# plot the second curve
fig.add_trace(go.Scatter(x=x2, y=y2, mode='lines', showlegend=False,
                                    line=dict(
                                        color='red',
                                        width=8)))
# plot the incremental steps
for i in range(N-1):
  fi = 2*(i+1)
  xx = lilbk1x[fi-2:fi-1] # used to be [fi-1:fi+1]
  yy = lilbk1y[fi-2:fi-1]
  fig.add_trace(go.Scatter(x=xx, y=yy, mode='markers', showlegend=False, visible="legendonly",
                                    line=dict(
                                        color='white',
                                        width=8)))
  xx = lilbk2x[fi-1:fi+1]
  yy = lilbk2y[fi-1:fi+1]
  fig.add_trace(go.Scatter(x=xx, y=yy, mode='markers', showlegend=False, visible="legendonly",
                                    line=dict(
                                        color='white',
                                        width=8)))

steps = []
levels = range(N-1)
for i in range(len(levels)*2+2):
    step = dict(
        method="update",
        label=str(i-1), #this is the step label
        args=[{"visible": [False] * (len(levels)*2+2)}])
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    for k in range(i):
      step["args"][0]["visible"][k] = True # make all previous traces visible
    steps.append(step)


sliders = [dict(
    active=0,
    currentvalue={"prefix": "step: "},
    pad={"t": 50},
    steps=steps[2:]
)]

fig.update_layout(sliders=sliders,
                  template="seaborn",
                  yaxis=dict(scaleanchor="y", scaleratio=1,
                             visible=False),
                  xaxis=dict(visible=False)


)

fig.update_xaxes(range=[0,12])
fig.update_yaxes(range=[0.6, 2.2])

fig.data[2].visible = True
fig.data[3].visible = True

fig.show()

fig.write_html("week2_two_parameterizations_of_curve.html")
```
