```
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

# Directional Derivative

```
# data for surface
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x)*np.arctan(y)

# x-y plane
z2 = x*y*0

# line on x-y plane
x3 = np.linspace(0.5, 2,100)
y3 = np.linspace(0.5, 2, 100)
z3 = x3*y3*0

# dot on the surface
x33 = [1.3]
y33 = [1.3]
z33 = np.sin(x33)*np.arctan(y33)

# line on the surface
z4 = np.sin(x3)*np.arctan(y3)

# tangent line at x=1.3, y=1.3 (equation taken from tangent plane equation)
i=1.3
z5 = np.sin(i)*np.arctan(i)+np.cos(i)*np.arctan(i)*(x3-i)+np.sin(i)*(y3-i)/(i**2+1)

# Initialize figure
fig = go.Figure()

# Add Traces
fig.add_trace(
    go.Surface(z=z, x=x, y=y, showscale=False, visible = "legendonly",
               name="Surface"))

fig.add_trace(go.Surface(x=x, y=y, z=z2, showscale=False, visible = "legendonly",
                         colorscale = ["tan", "blue"], opacity=0.5))
# line on x-y plane
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z3, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='darkblue',
                                        width=3)))
# dot on x-y plane
fig.add_trace(go.Scatter3d(x=x33, y=y33, z=[0], mode="markers", showlegend=False,
                           visible = "legendonly"))
# curve on surface
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z4, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='green',
                                        width=3)))
# dot on surface
fig.add_trace(go.Scatter3d(x=x33, y=y33, z=z33, mode="markers", showlegend=False,
                           visible = "legendonly"))
# tangent line
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z5, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='blue',
                                        width=3)))

# another line on x-y plane
x3 = np.linspace(-1.5, -0.5, 100)
y3 = x3*0.5-1
z3 = x3*y3*0

# another point on surface
x33 = [-1]
y33 = [-1*0.5-1]
z33 = np.sin(x33)*np.arctan(y33)

# another line on the surface
z4 = np.sin(x3)*np.arctan(y3)

# tangent plane at x = -0.75, y = -1.8
a=-1
b=-1*0.5-1
# z8 = np.cos(a)*np.arctan(b)*(x-a)+np.sin(a)*(1/(1+b**2))*(y-b)+np.sin(a)*np.arctan(b)

# tangent line along plane for constant y
x5 = x3
y5 = y3
z5 = np.cos(a)*np.arctan(b)*(x5-a)+np.sin(a)*(1/(1+b**2))*(y5-b)+np.sin(a)*np.arctan(b)

# Add More Traces
# line on x-y plane
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z3, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='darkblue',
                                        width=3)))
# dot on x-y plane
fig.add_trace(go.Scatter3d(x=x33, y=y33, z=[0], mode="markers", showlegend=False,
                           visible = "legendonly"))
# curve on surface
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z4, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='green',
                                        width=3)))
# dot on surface
fig.add_trace(go.Scatter3d(x=x33, y=y33, z=z33, mode="markers", showlegend=False,
                           visible = "legendonly"))
# tangent line
fig.add_trace(go.Scatter3d(x=x5, y=y5, z=z5, mode='lines', showlegend=False,
                           visible = "legendonly",
                                    line=dict(
                                        color='blue',
                                        width=3)))

fig.update_layout(template="seaborn",
                  scene = dict(aspectratio=dict(x=1, y=1, z=0.60),
                  xaxis=dict(range=[-2.5,2.5]),
                  yaxis=dict(range=[-2.5,2.5]),
                  zaxis=dict(range=[-1.5, 1.5])),
    updatemenus=[
        dict(
            type="buttons",
            #direction="right",
            active=0,
            #x=0.57,
            #y=1.2,
            buttons=list([
                dict(label="Surface",
                     method="update",
                     args=[{"visible": [True, False, False, False, False, False, False, False, False, False, False, False]}]),
                dict(label="x-y plane",
                     method="update",
                     args=[{"visible": [True, True, False, False, False, False, False, False, False, False, False, False]}]),
                dict(label="line 1 on x-y plane",
                     method="update",
                     args=[{"visible": [True, True, True, True, False, False, False, False, False, False, False, False]}]),
                dict(label="curve 1 on surface",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True, False, False, False, False, False, False]}]),
                dict(label="tangent line 1",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True, True, False, False, False, False, False]}]),
                dict(label="line 2 on x-y plane",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True, True, True, True, False, False, False]}]),
                dict(label="curve 2 on surface",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True, True, True, True, True, True, False]}]),
                dict(label="tangent line 2",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True, True, True, True, True, True, True]}])
            ]),
        )
    ])

fig.show()

fig.write_html("week5_directional_derivative.html")
```
