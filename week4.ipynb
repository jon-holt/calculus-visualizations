```
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

# Partial derivative and tangent plane

```
# data for surface
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x)*np.arctan(y)

# data for tangents (d/dx)
x2 = np.linspace(-2,2,100)
y2 = np.linspace(1.3, 1.3, 100)
b = np.sin(1.3)*np.arctan(1.3)-np.arctan(1.3)*np.cos(1.3)*1.3
z2 = np.arctan(1.3)*np.cos(1.3)*x2 + b # y = mx+b
z22 = np.sin(1.3)*np.arctan(1.3)+np.cos(1.3)*np.arctan(1.3)*(x2-1.3) # simplified

# data for tangents (d/dy)
x3 = np.linspace(1.3, 1.3, 100)
y3 = np.linspace(-2,2,100)
b = np.sin(1.3)*np.arctan(1.3)-np.sin(1.3)/(1+1.3**2)*1.3
z3 = np.sin(1.3)*1/(1+1.3**2)*y3 + b # y = mx+b
z33 = np.sin(1.3)*np.arctan(1.3)+np.sin(1.3)*(y3-1.3)/(1.3**2+1) # simplified

# data for plane
z4 = np.sin(1.3)*np.arctan(1.3)+np.cos(1.3)*np.arctan(1.3)*(x-1.3)+np.sin(1.3)*(y-1.3)/(1.3**2+1)

# Initialize figure
fig = go.Figure()

# Add Traces

fig.add_trace(
    go.Surface(z=z, x=x, y=y, showscale=False, visible = "legendonly",
               name="Surface"))

fig.add_trace(go.Mesh3d(x=[2,2,-2,-2], y=[1.3,1.3,1.3,1.3], z=[-1,1,1,-1], delaunayaxis='y', visible = "legendonly", opacity=0.7))

fig.add_trace(go.Mesh3d(x=[1.3,1.3,1.3,1.3], y=[2,2,-2,-2], z=[-1,1,1,-1], delaunayaxis='x', visible = "legendonly", opacity=0.7))

fig.add_trace(go.Scatter3d(x=x2, y=y2, z=z22, mode='lines', showlegend=False, visible = "legendonly",
                                    line=dict(
                                        color='darkblue',
                                        width=2)))

fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z33, mode='lines', showlegend=False, visible = "legendonly",
                                   line=dict(
                                        color='darkblue',
                                        width=2)))

fig.add_trace(go.Surface(x=x, y=y, z=z4, showscale=False, visible = "legendonly", colorscale = ["tan", "blue"]))

fig.update_layout(template="seaborn",
                  scene = dict(aspectratio=dict(x=1, y=1, z=0.60),
                  xaxis=dict(range=[-2.5,2.5]),
                  yaxis=dict(range=[-2.5,2.5]),
                  zaxis=dict(range=[-1.5, 1.5])),
    updatemenus=[
        dict(
            type="buttons",
            #direction="right",
            active=0,
            #x=1.0,
            #y=1.2,
            buttons=list([
                dict(label="Surface",
                     method="update",
                     args=[{"visible": [True, False, False, False, False, False]}]),
                dict(label="Transect x",
                     method="update",
                     args=[{"visible": [True, True, False, False, False, False]}]),
                dict(label="Transect y",
                     method="update",
                     args=[{"visible": [True, True, True, False, False, False]}]),
                dict(label="Tangent line x",
                     method="update",
                     args=[{"visible": [True, True, True, True, False, False]}]),
                dict(label="Tangent line y",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, False]}]),
                dict(label="Tangent plane",
                     method="update",
                     args=[{"visible": [True, True, True, True, True, True]}]),
                dict(label="Plane simple",
                     method="update",
                     args=[{"visible": [True, False, False, True, True, True]}]),
            ]),
        )
    ])

fig.show()

fig.write_html("week4_partial_derivative_tangent_plane.html")
```

# Differentiability (lack of)

```
# data for surface
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
z = x*y/np.sqrt(x**2+y**2)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, showscale=False)])

fig.update_layout(template="seaborn")
fig.show()

fig.write_html("week4_differentiability.html")
```
